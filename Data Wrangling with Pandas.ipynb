# Limpieza

```python
# Seteo la columna ID del DF unificado como Index
df.set_index('ID', drop=True, inplace=True)
```

```python
# Chequeo tipo de datos de las columnas del DF
df.dtypes
```

```python
# Chequeo valores de la columna 'Target' del DF
df.Target.unique()
```

```python
# Chequeo cantidad de nulos
df.isnull().sum()
```

```python
# Visualizo el registro nulo
df[(df.Target.isnull()==True)]
```

```python
# Elimino nulos al ser una única fila
df.dropna(inplace=True)
df.shape
```

```python
# Reduzco los valores de la columna Target a solo dos valores distintos utilizando el método .replace()
df["Target"]= df["Target"].replace([" <=50K.", " >50K."],[" <=50K", " >50K"])
df.Target.unique()
```

```python
# Otra forma de hacerlo: con el método .loc
df.loc[df.Target==' >50K.','Target']=' >50K'
df.loc[df.Target==' <=50K.','Target']=' <=50K'
df.Target.unique()
```

```python
# Otra alternativa: indicando el tipo de dato del dato a reemplazar
# df["Target"]= df["Target"].str.replace("<=50K.",  "<=50K")
# df["Target"]= df["Target"].str.replace(">50K.",  ">50K")
```

```python
# Modifico a categórico el tipo de datos de las columnas Target, Sex, Race, Country, Relationship y Workclass
df['Target']= df['Target'].astype('category')
df['Sex']= df['Sex'].astype('category')
df['Race']= df['Race'].astype('category')
df['Country']= df['Country'].astype('category')
df['Relationship']= df['Relationship'].astype('category')
df['Workclass']= df['Workclass'].astype('category')
df.dtypes
```

# Análisis

```python
# Proporciones 'Target' - 'Sex'
# Asigno el conteo de la agrupación por Target y Sexo a una variable
df_target_sex = df.groupby(by= ['Target','Sex']).count()
df_target_sex
```

```python
# Agrupo por el primer nivel del multi-índice (Level 0 = 'Target') e itero los valores de este en el índice de segundo nivel 
# Divido cada uno de los valores del índice de segundo nivel por el total del índice de primer nivel al que pertenecen
# Por último, multiplico por 100 para mostrarlo en porcentaje
df_target_sex = df_target_sex.groupby(level=[0]).apply(lambda x: x / x.sum() * 100).round(2)
df_target_sex
```

```python
# Limpio el df_target_sex para una mejor visualización de los datos
df_target_sex.drop(['Race','Country','Education','Education-Num','Marital Status','Relationship','Workclass','fnlwgt','Capital Gain','Capital Loss','Hours per week'], axis='columns', inplace=True)
df_target_sex.rename(columns={'Age':'Percentage'}, inplace=True)
df_target_sex
```

```python
# Proporciones 'Target' - 'Race'
df_target_race = df.groupby(by= ['Target','Race']).count()
df_target_race = df_target_race.groupby(level=[0]).apply(lambda x: x / x.sum() * 100).round(2)
df_target_race.drop(['Sex','Country','Education','Education-Num','Marital Status','Relationship','Workclass','fnlwgt','Capital Gain','Capital Loss','Hours per week'], axis='columns', inplace=True)
df_target_race.rename(columns={'Age':'Percentage'}, inplace=True)
df_target_race
```

```python
# Proporciones 'Target' - 'Education' (utilizando value_counts)
df_target_education2 = df.groupby(by=['Target'])
df_target_education2.Education.value_counts(normalize=True) * 100
```

```python
# Proporciones 'Target' - 'Education'
df_target_education = df.groupby(by= ['Target','Education']).count()
df_target_education = df_target_education.groupby(level=[0]).apply(lambda x: x / x.sum() * 100).round(2)
df_target_education.drop(['Sex','Race','Country','Education-Num','Marital Status','Relationship','Workclass','fnlwgt','Capital Gain','Capital Loss','Hours per week'], axis='columns', inplace=True)
df_target_education.rename(columns={'Age':'Percentage'}, inplace=True)
df_target_education
```
